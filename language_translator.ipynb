```python
from googletrans import Translator, LANGUAGES

def translate_text(input_text, from_lang, to_lang):
    translator = Translator()
    from_lang_key = list(LANGUAGES.keys())[list(LANGUAGES.values()).index(from_lang)]
    to_lang_key = list(LANGUAGES.keys())[list(LANGUAGES.values()).index(to_lang)]

    translation = translator.translate(input_text, src=from_lang_key, dest=to_lang_key)
    return translation.text

def main():
    print("Language Translator")
    languages = list(LANGUAGES.values())

    print("Available languages:")
    for lang in languages:
        print(lang)

    from_lang = input("Enter source language: ")
    to_lang = input("Enter target language: ")
    input_text = input("Enter text to translate: ")

    if from_lang not in languages or to_lang not in languages:
        print("Invalid language selection")
        return

    translated_text = translate_text(input_text, from_lang, to_lang)
    print("Translated text:", translated_text)

if __name__ == "__main__":
    main()
```
